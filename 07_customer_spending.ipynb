```python

from pyspark import SparkConf, SparkContext
```

```python

conf = SparkConf().setMaster('local').setAppName('customer_spend_analysis')
sc = SparkContext(conf = conf)
```

```python
def line_parser(line):
    fields = line.split(',')
    customer = int(fields[0])
    item = int(fields[1])
    price = float(fields[2])
    return customer, item, price
```

```python

lines = sc.textFile('spark_resources/customer-orders.csv')
parsed_lines = lines.map(line_parser)
```

```python

items = parsed_lines.map(lambda x: (x[0], x[2]))
totals_by_customer = items.reduceByKey(lambda x,y: (x+y))
```

```python

results = totals_by_customer.collect()

for result in results:
    print (result)
```
