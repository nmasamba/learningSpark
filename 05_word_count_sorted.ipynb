```python

import re
from pyspark import SparkConf, SparkContext
```

```python

def normalise_words(text):
    return re.compile('r\W+', re.UNICODE).split(text.lower())
```

```python

conf = SparkConf().setMaster('local').setAppName('sorted_word_count')
sc = SparkContext(conf = conf)
```

```python

input = sc.textFile('spark_resources/book.txt')
words = input.flatMap(normalise_words)
```

```python

word_counts = words.map(lambda x: (x,1)).reduceByKey(lambda x, y: x + y)
word_counts_sorted = word_counts.map(lambda (x,y): (y,x)).sortByKey()
```

```python

results = word_counts_sorted.collect()
```

```python

for result in results:
    count = str(result[0])
    word = result[1].encode('ascii', 'ignore')
    if (word):
        print (word) + ":\t\t" + count
```
